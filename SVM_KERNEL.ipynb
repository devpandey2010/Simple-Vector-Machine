```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings("ignore")
```

```python
x=np.linspace(-6,6,100)
y=np.sqrt(10**2-x**2)
y=np.hstack([y,-y])
x=np.hstack([x,-x])
```

```python
x
```

```python
y
```

```python
x1=np.linspace(-6,6,100)
y1=np.sqrt(4**2-x1**2)
y1=np.hstack([y1,-y1])
x1=np.hstack([x1,-x1])
```

```python
x1,y1
```

```python
#why hstack is used
plt.scatter(y,x)
plt.scatter(y1,x1)
```

```python
#hastack means you have some positive data points say now you want correspong negative data points
#here we hsatcked y,-y....x,-x
```

```python
df1=pd.DataFrame(np.vstack([y,x]).T,columns=['x1','x2'])
df1['Y']=0
df1
```

```python
df2=pd.DataFrame(np.vstack([y1,x1]).T,columns=['x1','x2'])
df2['Y']=1
df2
```

```python
df=pd.concat([df1,df2],ignore_index=True)
```

```python
df
```

```python
df1=df.copy()
df1
```

```python
df1["x1square"]=df1["x1"]**2
df1["x2square"]=df1["x2"]**2
df1["x1x2"]=df1["x1"]*df1["x2"]
df1.head()

```

```python
df1=df1.dropna()
df1
```

```python
x=df1[["x1square","x2square","x1x2"]]
y=df1["Y"]
```

```python
x
```

```python
#tarin test split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=2)
```

```python
import plotly.express as px
px.scatter_3d(df1,x="x1square",y="x2square",z="x1x2",color="Y")
```

```python
#we have already created polynomial from x1,x2 to x1square ,x2square,x1x2
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
model=SVC(kernel="linear")
model
```

```python
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
accuracy_score(y_test,y_pred)
```

```python
df=df.dropna()
df
```

```python
x=df[["x1","x2"]]
y=df["Y"]
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=3)
```

```python
#we wiil do it directly with sklearn
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
model1=SVC(kernel="poly")
model1.fit(x_train,y_train)
```

```python
y_pred=model1.predict(x_test)
```

```python
accuracy_score(y_pred,y_test)
```

```python
#poly kernel didinot perform well,so we wiil use rbf as our data is circular type
```

```python
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
model1=SVC(kernel="rbf")
model1.fit(x_train,y_train)
```

```python
y_pred=model1.predict(x_test)
```

```python
accuracy_score(y_pred,y_test)
```

```python
#lets see  for sigmoid as well
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
model1=SVC(kernel="sigmoid")
model1.fit(x_train,y_train)
```

```python
y_pred=model1.predict(x_test)
```

```python
accuracy_score(y_pred,y_test)
```

```python
#hyper parameter tuning
params={"kernel":["linear","poly","sigmoid","rbf"],
        "C":[0.1,0.2,1.0,2.0],
        "degree":[1,2,3,4,5],
```

```python
from sklearn.model_selection import GridSearchCV
newmodel=GridSearchCV(model1,param_grid=params,cv=5,verbose=3)
```

```python
newmodel.fit(x_train,y_train)
```

```python
newmodel.best_estimator_
```

```python
newmodel.best_params_
```

```python
newmodel.best_score_
```

```python
accuracy_score(y_test,newmodel.best_estimator_.predict(x_test))
```
